```python
%matplotlib inline
import matplotlib.pyplot as plt
from detdata import DetGen
from detdata.augmenters import crazy_augmenter
import imgaug


detg= DetGen('/home/i008/malaria_data/dataset_train.mxrecords',
      '/home/i008/malaria_data/dataset_train.csv',
      '/home/i008/malaria_data/dataset_valid.mxindex', batch_size=16)


aug_gen =detg.get_augmenting_generator(augmenter=crazy_augmenter)

images, boxes, classes = next(aug_gen)

collect_images = []
for b, i in zip(boxes, images):
    collect_images.append(b.draw_on_image(i, thickness=4))
    
plt.figure(figsize=(13,13))
plt.imshow(imgaug.draw_grid(collect_images))
```
